```python
#Limpieza de columnas con numerosos NaN
ensanut_clean = ensanut_f_d_rename.drop(['papanicolaou', 'papiloma', 'tuberculosis', 'fumó_pasado', 'bebe_frecuencia'], axis=1) 
ensanut_clean.head()
```

```python
#Rellenar NaN con 8=No responde
ensanut_clean['exploración_senos'] = ensanut_clean['exploración_senos'].fillna(8)
```

```python
ensanut_clean['cáncer_mama'] = ensanut_clean['cáncer_mama'].fillna(8)
ensanut_clean['cigarro_electrónico'] = ensanut_clean['cigarro_electrónico'].fillna(8)
```

```python
#NaN por columna
ensanut_clean.isna().sum(axis=0)
```

```python
ensanut_clean
```

```python
#Filtrar 8=No responde 9=No sabe en pregunta detección de cáncer de mama
filtro_cancer = ensanut_clean['cáncer_mama'] < 3
ensanut_cancer = ensanut_clean[filtro_cancer]
```
